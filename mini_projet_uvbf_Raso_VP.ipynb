 3.1: Import des bibiliothèques

```python
import os
import pandas as pd
import string
import xlsxwriter
```

3.2 Chargement et préparation des données

```python
#3.2.1 Indexation du repertoire de travail
os.chdir("C:/Users/HP/Desktop/Code_Jupiter")

#3.2.2 Nomination des colonnes
items=["geonameid", "name", "asciiname", "alternatenames", "latitude", "longitude", "feature class", "feature code", "country code", 
     "cc2", "admin1 code", "admin2 code", "admin3 code", "admin4 code", "population", "elevation", "dem", "timezone","modification date"]

#1.3 Chargement de la base de données
bfa = pd.read_csv("BF.txt", delimiter="\t", header=None, names=items, keep_default_na=False)
bfa #Affichage de la base
```

```python
#3.2.3. Préparation de la base de données

items2 = ["geonameid", "asciiname", "latitude", "longitude"]

extract_bfa = bfa[items2]

# 3.2.4. Rénommination des colonnes

rename_items2 = {"geonameid":"ID", "asciiname":"location_name", "latitude":"lat", "longitude":"long"}

extract_bfa.rename(columns = rename_items2, inplace=True)
```

```python
#3.2.5. Enregitstrement de l base de données au format csv

extract_bfa.to_csv("burkina_location.csv")
```

4. Opérations sur fichier csv

```python
# 4.1. Chargement du fichier csv

burkina_loc = pd.read_csv('burkina_location.csv')
```

```python
# 4.2. Extraction des données contenant le nom gounghin

burkina_loc_gounghin = burkina_loc[burkina_loc.location_name.str.lower()=="gounghin"]
```

```python
# 4.3. Enregistrement de burkina_location_gounghin sous format csv

burkina_loc_gounghin.to_csv('gounghin.csv')
```

```python
# 4.4. Extraction des lieux dont les noms sont compris entre A et P

a_p = list(string.ascii_uppercase[:16])

burkina_loc_a_p = burkina_loc[burkina_loc.location_name.str[0].isin(a_p)].sort_values(by='location_name')
```

```python
# 4.5. Enregistrement de burkina_location_a_p sous format csv

burkina_loc_a_p.to_csv('burkina_location_a_p.csv')
```

```python
# 4.6. Affichage des latitudes minimales de burkina_location.csv

burkina_loc[burkina_loc.lat == burkina_loc.lat.min()]
```

```python
# 4.7. Affichage des latitudes minimales de burkina_location.csv

burkina_loc[burkina_loc.long == burkina_loc.long.min()]
```

```python
# 4.8. Les leiux dont lat>=11 et long<=0.5

burkina_loc[(burkina_loc.lat >= 11 ) & (burkina_loc.long <= 0.5 ) ]
```

3.Sortie Excel

```python
# Création du fichier Excel nommé mini_projet

book  = xlsxwriter.Workbook(os.path.join(os.getcwd(), 'mini_projet.xlsx'))
sheet_gounghin = book.add_worksheet('gounghin')
sheet_a_to_p = book.add_worksheet('A_to_P')

colonnes = [c for c in list(burkina_loc_gounghin.columns) if c!='Unnamed: 0']

# Création des entêtes 

for i in range(len(colonnes)):
        sheet_gounghin.write(0, i, colonnes[i])
        sheet_a_to_p.write(0, i, colonnes[i])
        
# Renseignement des données

for i in range(burkina_loc_gounghin.shape[0]):
    for j in range(len(colonnes)):
        sheet_gounghin.write(i+1, j, burkina_loc_gounghin.iloc[i][colonnes[j]])

for i in range(burkina_loc_a_p.shape[0]):
    for j in range(len(colonnes)):
        sheet_a_to_p.write(i+1, j, burkina_loc_a_p.iloc[i][colonnes[j]])
book.close()
```
